```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
# Equação do pêndulo: d^2theta/dt^2 = -(g/L) * sin(theta)
def pendulum(state, t, g, L):
    theta, omega = state  # theta é o ângulo, omega é a velocidade angular
    dtheta_dt = omega
    domega_dt = -(g / L) * np.sin(theta)
    return [dtheta_dt, domega_dt]
```

```python
# Parâmetros
g = 9.81  # Gravidade (m/s^2)
L = 1.0   # Comprimento do pêndulo (m)
t = np.linspace(0, 10, 1000)  # Tempo
initial_state = [np.pi / 4, 0]  # Ângulo inicial de 45° e velocidade inicial 0
```

```python
# Resolver a equação diferencial
solution = odeint(pendulum, initial_state, t, args=(g, L))
theta = solution[:, 0]
```

```python
plt.figure(figsize=(12, 7))
plt.plot(t, theta, label="Ângulo (θ)")
plt.title("Oscilações de um Pêndulo Simples", fontsize=16)
plt.xlabel("Tempo (s)", fontsize=12)
plt.ylabel("Ângulo (rad)", fontsize=12)
plt.grid(True)
plt.legend()
plt.show()
```
