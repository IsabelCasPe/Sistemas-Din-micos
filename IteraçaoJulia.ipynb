Calcula a trajetória de um ponto inicial z_0
 sob a iteração z_{n+1} = z_n^2 + c
Se o módulo de ( z ) ultrapassar 2, o ponto "escapa" e paramos.



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Função que calcula a iteração do Conjunto de Julia para um ponto z
def iterate_julia(z, c, max_iter=100):
    trajectory = [z]  # Lista para armazenar a trajetória do ponto
    for _ in range(max_iter):
        z = z**2 + c  # Iteração: z_{n+1} = z_n^2 + c
        trajectory.append(z)
        if abs(z) > 2:  # Se o ponto "escapa", paramos
            break
    return np.array(trajectory)
```

```python
# Configurações do gráfico
plt.figure(figsize=(7, 6))
plt.title("Trajetórias no Conjunto de Julia")
plt.xlabel("Parte Real")
plt.ylabel("Parte Imaginária")
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
# Pontos iniciais para as trajetórias (você pode adicionar mais!)
initial_points = [
    0.1 + 0.1j,
    0.5 + 0.2j,
    -0.5 - 0.5j,
    0.1 + 0.5j
]

# Cores para cada trajetória
colors = ['blue', 'red', 'green', 'purple']

# Plotar as trajetórias
for z0, color in zip(initial_points, colors):
    trajectory = iterate_julia(z0, c)
    # Separar parte real e imaginária para o gráfico
    x = trajectory.real
    y = trajectory.imag
    # Plotar a linha da trajetória
    plt.plot(x, y, color=color, alpha=0.7, linewidth=1.5, label=f'z0 = {z0}')
    # Marcar o ponto inicial com um círculo
    plt.scatter(x[0], y[0], color=color, s=100, edgecolor='black', zorder=5)
    # Marcar o último ponto com um triângulo (se escapar ou acabar)
    plt.scatter(x[-1], y[-1], color=color, marker='^', s=100, edgecolor='black', zorder=5)

# Configurações estéticas do gráfico
plt.title(f"Trajetórias no Conjunto de Julia (c = {c})", fontsize=16, pad=20)
plt.xlabel("Parte Real", fontsize=12)
plt.ylabel("Parte Imaginária", fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.5)
plt.axhline(0, color='black', linewidth=0.5, alpha=0.3)
plt.axvline(0, color='black', linewidth=0.5, alpha=0.3)

# Ajustar os limites do gráfico para melhor visualização
plt.xlim(-3, 3)
plt.ylim(-3, 3)

# Mostrar o gráfico
plt.show()
```

Cada trajetória é plotada com uma cor diferente, com o ponto inicial marcado por um círculo e o final por um triângulo. A legenda ajuda a identificar cada ponto inicial.
