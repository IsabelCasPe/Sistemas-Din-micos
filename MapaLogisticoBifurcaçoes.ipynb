 #exemplo de bifurcação em sistemas dinâmicos. A bifurcação ocorre quando uma pequena mudança em um parâmetro do sistema causa uma mudança qualitativa no comportamento das soluções de longo prazo.

Para ilustrar, podemos usar o Mapa Logístico. Esse sistema simples é dado pela fórmula:
x_(n+1)=r⋅x_n⋅(1-x n)

onde:
x_n representa a população em uma geração n,

r é um parâmetro que controla o crescimento populacional.
O comportamento do sistema muda radicalmente dependendo do valor de
r. Com valores baixos de r, a população se estabiliza, mas conforme r aumenta, a população passa a oscilar, eventualmente entrando em comportamento caótico.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Intervalo de valores de r (taxa de crescimento) e configurações do sistema
r_values = np.linspace(2.5, 4.0, 10000)  # Valores de r para explorar o comportamento dinâmico
iterations = 1000                        # Número de iterações para cada valor de r
last = 100                               # Últimos pontos para visualização
```

```python
# Valor inicial pequeno para x (população)
x = 1e-5 * np.ones(len(r_values))
```

```python
# Listas para armazenar os resultados
r_plot = []
x_plot = []
```

```python
# Cálculo do Mapa Logístico para cada valor de r
for i in range(iterations):
    x = r_values * x * (1 - x)           # Aplicação da fórmula do Mapa Logístico
    if i >= (iterations - last):         # Armazena apenas as últimas iterações
        r_plot.extend(r_values)
        x_plot.extend(x)
```

```python
# Criação do diagrama de bifurcação
plt.figure(figsize=(12, 7))
plt.plot(r_plot, x_plot, ',k', alpha=0.25)  # Gráfico de dispersão das bifurcações
plt.xlim(2.5, 4.0)
plt.xlabel("Taxa de Crescimento (r)")
plt.ylabel("População Estável (x)")
plt.title("Diagrama de Bifurcação do Mapa Logístico")
plt.show()
```

```python
# Valores de r e número de iterações
r_values = np.linspace(1, 5.0, 100000)  # Intervalo de valores de r para explorar
iterations = 1000                        # Iterações para cada valor de r
last = 100                               # Últimos pontos para visualização

# Prepara o espaço para a população
x = 1e-5 * np.ones(len(r_values))        # Ponto inicial para x

# Listas para armazenar resultados
r_plot = []
x_plot = []

# Cálculo do Mapa Logístico para cada valor de r
for i in range(iterations):
    x = r_values * x * (1 - x)           # Aplica a função do Mapa Logístico
    if i >= (iterations - last):         # Armazena apenas as últimas iterações
        r_plot.extend(r_values)
        x_plot.extend(x)
```

```python
# Criação do gráfico de bifurcação
plt.figure(figsize=(12, 7))
plt.plot(r_plot, x_plot, ',k', alpha=0.25)  # Gráfico de dispersão para bifurcações
plt.xlim(2.5, 4.0)
plt.xlabel("Taxa de Crescimento (r)")
plt.ylabel("População Estável (x)")
plt.title("Diagrama de Bifurcação do Mapa Logístico")
plt.show()
```

 # Mapa Logístico. Ele mostra como o sistema passa de comportamento estável para caótico à medida que o parâmetro r aumenta.

O diagrama de bifurcação do Mapa Logístico é um excelente exemplo visual de como sistemas dinâmicos simples podem transitar do comportamento previsível para o caótico.

# Interpretação do Diagrama de Bifurcação:
Valores de r Baixos (até cerca de 3):

Quando r á entre 2.5 e 3, o sistema converge para um ponto fixo estável. Isso significa que, independentemente da condição inicial, a população acaba se estabilizando em um único valor. Aqui, o comportamento é previsível e não caótico.
Oscilações Periódicas (Entre r≈3 e r≈3.5):

Conforme r aumenta além de 3, o sistema passa por uma bifurcação. Em vez de um único ponto de equilíbrio, a população começa a oscilar entre dois valores (uma órbita periódica de período 2).
À medida que r continua a aumentar, o número de pontos na órbita periódica dobra repetidamente (período 4, depois 8, e assim por diante). Esse processo é conhecido como rota para o caos por duplicação de período.
Transição para o Caos (Valores de r acima de 3.57):

Quando  r ultrapassa um certo limite crítico (cerca de 3.57), o sistema entra em um regime caótico. Aqui, o valor da população varia de maneira aparentemente aleatória e imprevisível, sem repetir um ciclo fixo.
No caos, pequenas diferenças nas condições iniciais resultam em trajetórias de longo prazo drasticamente diferentes, ilustrando a sensibilidade a condições iniciais.
Janelas de Ordem (Ex.:r≈3.83):

Apesar do comportamento caótico predominante para valores altos de
r, há pequenas janelas de ordem onde o sistema volta a exibir órbitas periódicas. Isso pode ser visto como "ilhas" de estabilidade em meio ao caos.
#Conclusão
O diagrama de bifurcação do Mapa Logístico mostra a complexidade dos sistemas dinâmicos não lineares e como mudanças em um parâmetro podem levar a comportamentos drasticamente diferentes. Esse fenômeno é muito relevante em ecologia, economia, clima e outros sistemas onde o caos pode surgir de equações relativamente simples.
