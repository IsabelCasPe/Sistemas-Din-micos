```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.integrate import solve_ivp
from matplotlib.animation import FuncAnimation
```

```python
# Par√¢metros do sistema de Lorenz
sigma = 10.0
rho = 28.0
beta = 8.0 / 3.0
```

```python
# Fun√ß√£o que define o sistema de Lorenz
def lorenz(t, state):
    x, y, z = state
    dx_dt = sigma * (y - x)
    dy_dt = x * (rho - z) - y
    dz_dt = x * y - beta * z
    return [dx_dt, dy_dt, dz_dt]
```

```python
# Condi√ß√µes iniciais
initial_state = [0.0, 1.0, 1.05]

# Tempo de integra√ß√£o
t_span = (0, 40)
t_eval = np.linspace(t_span[0], t_span[1], 10000)

# Solu√ß√£o num√©rica do sistema
solution = solve_ivp(lorenz, t_span, initial_state, t_eval=t_eval)

# Extra√ß√£o das trajet√≥rias
x, y, z = solution.y
```

```python
# Visualiza√ß√£o do sistema de Lorenz em 3D
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, lw=0.5, color='b')
ax.set_title("Atra√ß√£o Ca√≥tica do Sistema de Lorenz")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
plt.show()
```
