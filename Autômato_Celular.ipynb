Autômatos celulares são sistemas dinâmicos discretos em grades, onde o estado de cada célula evolui com base em regras simples. A Regra 30 é um exemplo 1D caótico, enquanto o Jogo da Vida de Conway é um exemplo 2D famoso.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rule_30(grid):
    new_grid = np.zeros_like(grid)
    for i in range(1, len(grid) - 1):
        left, center, right = grid[i-1], grid[i], grid[i+1]
        new_grid[i] = left ^ (center | right)  # Regra 30: XOR de left com (center OR right)
    return new_grid
```

```python
# Inicializar grid 1D
width = 101
steps = 50
grid = np.zeros((steps, width), dtype=int)
grid[0, width // 2] = 1  # Começar com uma célula no meio
```

```python
# Evoluir
for t in range(1, steps):
    grid[t] = rule_30(grid[t-1])
```

```python
plt.figure(figsize=(12, 8))
plt.imshow(grid, cmap='binary')
plt.title("Autômato Celular - Regra 30", fontsize=16)
plt.xlabel("Posição")
plt.ylabel("Tempo")
plt.show()
```
