```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def logistic_map(r, x0, iterations):
    x = [x0]
    for _ in range(iterations - 1):
        x.append(r * x[-1] * (1 - x[-1]))
    return np.array(x)
```

```python
# Parâmetros
r = 3.0  # Taxa de crescimento
x0 = 0.5  # Condição inicial
iterations = 200
```

```python
# Calcular a trajetória
trajectory = logistic_map(r, x0, iterations)
```

```python
# Plotar
plt.figure(figsize=(10, 6))
plt.plot(trajectory, '-o', markersize=4, alpha=0.7)
plt.title(f"Mapa Logístico (r = {r})", fontsize=16)
plt.xlabel("Iteração", fontsize=12)
plt.ylabel("População (x)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
